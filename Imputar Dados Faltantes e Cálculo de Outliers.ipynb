O exercício foi feito com 4 objetivos:
1. Calcular e informar a média dos valores na coluna PESO. Em seguida, preencher essa coluna com o dado faltante.
2. Calcular e informar a mediana dos valores na coluna Altura. Em seguida, preencher essa coluna com o dado faltante.
3. Calcular a moda das colunas COR OLHO e COR CABELO. Em seguida, preencher essas colunas com os dados faltantes.
4. Utilizando o método Z-value, verificar se existem outliers na coluna IMC. Caso existam, apontar quais são.

```python
#importar bibliotecas
import pandas as pd
import numpy as np
```

```python
##criação do DataFrame
df = pd.read_csv('dados_biotipo.csv')
df
```

```python
df.info()
## avaliação dos tipos de dados no df indica a necessidade de tipar colunas identificadas como object que devem ser tratadas como float
## além disso, pode-se perceber que algumas células nas colunas Cor Cabelo e Cor olhos foram preenchidas apenas com " "
```

```python
#Alteração do tipo de dados nas colunas Altura e Peso
df['Altura'] = pd.to_numeric(df['Altura'], errors='coerce')
df['Peso'] = pd.to_numeric(df['Peso'], errors='coerce')

## descrição do dataframe e demonstração da média dos valores apenas nas colunas do tipo numeric.
df.describe()
```

```python
## preenchimento das células da coluna Peso com o valor da média.
df['Peso'] = df['Peso'].fillna(df['Peso'].mean())

## preenchimento das células da coluna Altura com o valor da mediana.
df['Altura'] = df['Altura'].fillna(df['Altura'].median())

## substituição do " " em células vazias nas colunas de Cor de Olho e Cabelo pela moda de cada respectiva coluna.
	##a função .mode() devolve um objeto do tipo DataFrame, então usei o iloc[0] para pegar apenas o valor da moda, ao invés do dataframe inteiro.
df['Cor Olho'] = df['Cor Olho'].replace(" ", df['Cor Olho'].mode().iloc[0])
df['Cor Cabelo'] = df['Cor Cabelo'].replace(" ", df['Cor Cabelo'].mode().iloc[0])
df
```

```python
## declaração da lista de outliers
outliers=[]

#função de detecção de outliers
def detect_outlier(data_1):
    threshold=3
    mean_1 = np.mean(data_1)
    std_1 =np.std(data_1)
    
    #calculo do z score de cada item na coluna IMC, inserção na lista de outliers se o score for maior que o threshold
    for y in data_1:
        z_score= (y - mean_1)/std_1 
        if np.abs(z_score) > threshold:
            outliers.append(y)
    return outliers

#chamar a função de detecão de outliers e fornecer como argumento a coluna de IMC do DataFrame
outlier_datapoints = detect_outlier(df['IMC'])
#imprimir a lista de outliers
print(outlier_datapoints)
```
